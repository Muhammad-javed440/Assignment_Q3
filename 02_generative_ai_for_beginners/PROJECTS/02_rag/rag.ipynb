```python

import pdfplumber

def split_into_chunks(text, words_per_chunk):
    words = text.split()
    return [" ".join(words[i:i + words_per_chunk]) for i in range(0, len(words), words_per_chunk)]

# Load the PDF and extract text
# Assuming 'uploaded' is a dictionary containing the uploaded PDF file
# You might need to adjust how you access the file path depending on
# the structure of 'uploaded'
file_path = list(uploaded.keys())[0]  # Get the file path from the dictionary
with pdfplumber.open(file_path) as pdf:
    full_text = " ".join(page.extract_text() for page in pdf.pages)

# Divide the text into chunks of 100 words each
chunks = split_into_chunks(full_text, 100)

# Print or process chunks
for i, chunk in enumerate(chunks):
    print(f"Chunk {i + 1}:")
    print(chunk)
    print("-" * 80)
```

```python
import chromadb
chroma_client = chromadb.Client()
```

```python
collection = chroma_client.get_or_create_collection(name="my_collection")
```

```python
collection.add(
    documents=[
        "This is a document about pineapple",
        "This is a document about oranges",
        "This is a document about apple",
        "This is a document about water melon",
        "This is a document about stabbery",
        "This is a document about banana",
        "This is a document about graps",
        "This is a document about melon",
        "This is a document about guava",
        "This is a document about punat",
    ],
    ids=["id1", "id2","id3", "id4","id5", "id6","id7", "id8","id9", "id10"]
)
```

```python
results = collection.query(
    query_texts=["This is a query document about apple"], # Chroma will embed this for you
    n_results=2 # how many results to return
)
print(results)
```
